# Charts in Matplotlis

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

print(__version__) # requires version >= 1.9.0
```

```python
import cufflinks as cf

# For Notebooks
init_notebook_mode(connected=True)

# For offline use
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
x = np.linspace(0,5,11)
y = x**2

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color = 'blue', linewidth= 2, alpha= 1, linestyle= '--', marker= 'o', markersize= 20
       , markerfacecolor= 'green', markeredgewidth= 3, markeredgecolor= 'yellow') 
```

# Charts in Seaborn

## Distribution Plot

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

##  Categorical Plot (Violin Plot)

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

##  Matrix Plot (Heat Map)

```python
pvflights = flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(pvflights)
```

## Regresssion Plot 

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
df3.iplot(kind='surface',colorscale='rdylbu')
```
